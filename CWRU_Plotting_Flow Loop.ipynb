```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

# === Base folder where the the excel file is kept ===
base_dir = r"C:\Users\mh272\Box\NED3_Share\Ishraq Hossain\CWRU\test17_17th Oct_Friday\Flow Loop Dataset\25gs_20C"  # Replace with the actual path
file_name = "25gs_20C.xlsx" # Replace with the actual name
file_path = os.path.join(base_dir, file_name)

# Make an output folder for images
images_dir = os.path.join(base_dir, "Plot Images")
os.makedirs(images_dir, exist_ok=True)

# Load the data from the Excel file
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Select rows from 1000 to 1200
#df_selected = df.iloc[0:1872]
df_selected = df.copy()

# Helper to make a safe filename from the plot title
def _safe_filename(title: str) -> str:
    # Remove illegal characters on Windows and strip extra spaces
    name = re.sub(r'[<>:"/\\|?*]+', '', title).strip()
    # Replace spaces with underscores
    name = re.sub(r'\s+', '_', name)
    return f"{name}.png"

# Define a function to plot and save each parameter with its average
def plot_with_average(data, column_name, title, ylabel, file_name):
    avg_value = data[column_name].mean()
    plt.figure()
    plt.plot(data.index, data[column_name], label=f'Average: {avg_value:.2f}')
    plt.title(title, fontsize = 24, fontname ='Arial')
    plt.xlabel('Row Index', fontsize = 20, fontname ='Arial')
    plt.ylabel(ylabel, fontsize = 20, fontname ='Arial')
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.legend()
    plt.tight_layout()
    #plt.show()
    # Save using the title as filename inside "Plot Images"
    out_path = os.path.join(images_dir, _safe_filename(title))
    plt.savefig(out_path, dpi=200)
    plt.show()
    plt.close()
    print(f" Saved: {out_path}")

# Plot flow rate with average
plot_with_average(df_selected, 'Mass Flow Rate [kg/s]', 'Flow Rate', 'Flow Rate [kg/s]', 'flow_rate_plot.png')

# Plot fluid velocity with average
plot_with_average(df_selected, 'Fluid Velocity [m/s]', 'Fluid Velocity', 'Velocity [m/s]', 'fluid_velocity_plot.png')

# Plot total heat flux with average
plot_with_average(df_selected, 'Total Heat Flux [W/cm]', 'Total Heat Flux', 'Heat Flux [W/cm]', 'total_heat_flux_plot.png')

# Plot inlet subcooling with average
plot_with_average(df_selected, 'Inlet Subcooling [C]', 'Inlet Subcooling', 'Subcooling [C]', 'inlet_subcooling_plot.png')

# Plot inlet pressure with average
plot_with_average(df_selected, 'Inlet Pressure [kPa]', 'Inlet Pressure', 'Pressure [kPa]', 'inlet_pressure_plot.png')

# Plot inlet temperature with average
plot_with_average(df_selected, 'Inlet Temperature [C]', 'Inlet Temperature', 'Temperature [C]', 'inlet_temperature_plot.png')

# Plot saturation temperature with average
plot_with_average(df_selected, 'Saturation Temperature Inlet [C]', 'Saturation Temperature', 'Temperature [C]', 'saturation_temperature_plot.png')

# Define the actual numerical positions for x1 to x7
positions = [0.0054, 0.0227, 0.04, 0.0573, 0.0746, 0.0919, 0.1092]

```

```python
# Plot average temperatures from x1 to x7 vs their numerical positions
temp_columns = [f'Temperature at x{i} [C]' for i in range(1, 8)]
avg_temps = df_selected[temp_columns].mean()
plt.figure()
plt.scatter(positions, avg_temps, marker='s', s=60, c='orange')
title = "Average Temperatures at x1 to x7"
plt.title(title, fontsize = 24, fontname ='Arial')
plt.xlabel('Position [m]', fontsize = 20, fontname ='Arial')
plt.ylabel('Average Temperature [C]', fontsize = 20, fontname ='Arial')
plt.tick_params(axis='both', which='major', labelsize=16)
#plt.savefig('average_temperature_plot.png')
out_path = os.path.join(images_dir, _safe_filename(title))
plt.savefig(out_path, dpi=200)
plt.show()
plt.close()


```

```python

# Plot average heat transfer coefficients from x1 to x7 vs their numerical positions
htc_columns = [f'Heat Transfer Coefficient at x{i} [W/mK]' for i in range(1, 8)]
avg_htc = df_selected[htc_columns].mean()
plt.figure()
plt.scatter(positions, avg_htc, marker='s', s=60, c='orange')
title = "Average Heat Transfer Coefficients at x1 to x7"
plt.title(title, fontsize = 24, fontname ='Arial')
plt.xlabel('Position [m]', fontsize = 20, fontname ='Arial')
plt.ylabel('Average Heat Transfer Coefficient [W/mK]', fontsize = 20, fontname ='Arial')
plt.tick_params(axis='both', which='major', labelsize=16)
#plt.savefig('average_temperature_plot.png')
out_path = os.path.join(images_dir, _safe_filename(title))
plt.savefig(out_path, dpi=200)
plt.show()
plt.close()


```
