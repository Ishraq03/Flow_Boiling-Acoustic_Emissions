```python
import re
import numpy as np
import pandas as pd
from datetime import datetime
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# --- 1) Pick the .txt file ---
Tk().withdraw()
file_path = askopenfilename(title="Select the dual-channel time .txt",
                            filetypes=[("Text files", "*.txt")])
assert file_path, "No file selected."
print(file_path)
```

```python
from datetime import datetime, timedelta  # <-- add this

def parse_dual_txt(file_path):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        lines = f.readlines()

    base_dt = datetime.strptime(lines[5].strip(), "%a %b %d %H:%M:%S %Y")

    t_list, para1_list, para2_list = [], [], []
    rms1, thr1, ae1, asl1 = [], [], [], []
    rms2, thr2, ae2, asl2 = [], [], [], []

    i, n = 0, len(lines)
    while i < n:
        line = lines[i]
        if line.startswith("  2 "):
            toks = line.split()
            t_sec = float(toks[1]); para1 = float(toks[2]); para2 = float(toks[3])
            if i + 2 >= n: break
            data_line = lines[i+2].strip()

            m1 = re.search(r"1:\[\s*([^\]]+)\]", data_line)
            m2 = re.search(r"2:\[\s*([^\]]+)\]", data_line)
            if not (m1 and m2):
                i += 1; continue

            def four_nums(block):
                a = block.split()
                return float(a[0]), float(a[1]), float(a[2]), float(a[3])

            r1, th1, abe1, a1 = four_nums(m1.group(1))
            r2, th2, abe2, a2 = four_nums(m2.group(1))

            t_list.append(t_sec)
            para1_list.append(para1); para2_list.append(para2)
            rms1.append(r1); thr1.append(th1); ae1.append(abe1); asl1.append(a1)
            rms2.append(r2); thr2.append(th2); ae2.append(abe2); asl2.append(a2)

            i += 3
        else:
            i += 1

    t0 = t_list[0]
    time_s = [t - t0 for t in t_list]

    time_hms = []
    for t in t_list:
        whole = int(t)
        frac7 = f"{int(round((t - whole) * 10_000_000)):07d}"  # <-- corrected 1e7
        tt = base_dt + timedelta(seconds=whole)
        time_hms.append(f"{tt.hour:02d}:{tt.minute:02d}:{tt.second:02d}.{frac7}")

    df_txt = pd.DataFrame({
        "Time [h,m,s]": time_hms,
        "Time [s]": time_s,
        "PARA_1": para1_list, "RMS_1": rms1, "THR_1": thr1, "ABS-ENERGY_1": ae1, "ASL_1": asl1,
        "PARA_2": para2_list, "RMS_2": rms2, "THR_2": thr2, "ABS-ENERGY_2": ae2, "ASL_2": asl2,
    })
    return df_txt, base_dt

# ---- usage ----
# file_path = r"C:\path\to\5gs_22C.DTA"
df_txt, base_dt = parse_dual_txt(file_path)
print(base_dt.strftime("%Y-%m-%d %H:%M:%S"))   # e.g., 2025-10-17 12:41:54
print(df_txt.head())
```

```python
def parse_hms_7digits(s):
    # s like "12:41:54.0100292"
    if "." in s:
        hms, frac = s.split(".")
        frac = frac[:6]  # keep only first 6 digits
        s6 = f"{hms}.{frac}"
        return datetime.strptime(s6, "%H:%M:%S.%f")
    else:
        return datetime.strptime(s, "%H:%M:%S")
    

ts1 = parse_hms_7digits(df_txt["Time [h,m,s]"][0])
ts2 = parse_hms_7digits(t_Time[0])

diff_seconds = (ts1 - ts2).total_seconds()
print(diff_seconds)
```
